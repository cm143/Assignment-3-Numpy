```python
import numpy as np
basketball_data = np.loadtxt('NBA_Player_Stats.tsv', delimiter = '\t', skiprows = 1, usecols = range(5,22))
player_data = np.loadtxt('NBA_Player_Stats.tsv', delimiter = '\t', skiprows = 1, usecols = [1,3], dtype = 'str')

fg_acc = basketball_data[:,2] / basketball_data[:,3]
tp_acc = basketball_data[:,4] / basketball_data[:,5]
ft_acc = basketball_data[:,6] / basketball_data[:,7]
avg_point = basketball_data[:,16] / basketball_data[:,1]
accuracy = fg_acc + tp_acc + ft_acc
blocks = basketball_data[:,15] / basketball_data[:,0]
steals = basketball_data[:,14] / basketball_data[:,0]
```

```python
#top 100 players field goal accuracy#
np.sort(fg_acc)[::-1][2:103]
fgacc_sort = np.argsort(fg_acc, axis = 0)[::-1][2:]
(player_data[fgacc_sort[:100]])
```

```python
#top 100 players three pointers accuracy#
np.sort(tp_acc)[::-1][4:103]
tpacc_sort = np.argsort(fg_acc, axis = 0)[::-1][2:]
(player_data[tpacc_sort[:100]])
```

```python
#top 100 players free throw accuracy#
np.sort(ft_acc)[::-1][2:103]
ftacc_sort = np.argsort(ft_acc, axis = 0)[::-1][2:]
(player_data[ftacc_sort[:100]])
```

```python
#top 100 players average points#
np.sort(avg_point)[::-1][2:103]
avgpoint_sort = np.argsort(avg_point, axis = 0)[::-1][2:]
(player_data[avgpoint_sort[:100]])
```

```python
#top 100 players overall accuracy#
np.sort(accuracy)[::-1][2:103]
accuracy_sort = np.argsort(accuracy, axis = 0)[::-1][2:]
(player_data[accuracy_sort[:100]])
```

```python
#top 100 players blocks per game#
np.sort(blocks)[::-1][2:103]
blocks_sort = np.argsort(blocks, axis = 0)[::-1][2:]
(player_data[blocks_sort[:100]])
```

```python
#top 100 players steals per game#
np.sort(steals)[::-1][2:103]
steals_sort = np.argsort(steals, axis = 0)[::-1][2:]
(player_data[steals_sort[:100]])
```
